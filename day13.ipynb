```python
# Learn polars
import polars as p1
df = p1.DataFrame({
    "name":["aman","suman","Rohit","henry","kamal"],
    "age":[23,24,24,21,36],
    "marks":[90,60,70,80,95],
    "salary":[10000,20000,30000,40000,50000]
})  

print(df)
```

```python
import polars as pl

df = pl.DataFrame({
    "Employee": ["Aman", "Rahul", "Priya", "Neha", "Rohit"],
    "Age": [22, 25, 20, 28, 24],
    "Department": ["IT", "HR", "IT", "Sales", "HR"],
    "Salary": [40000, 60000, 35000, 80000, 50000]
})

print(df)

it_dept = df.filter(pl.col("Department") == "IT")
print(it_dept)

high_salary = df.filter(pl.col("Salary") >= 50000)
print(high_salary)

df_bonus = df.with_columns((pl.col("Salary") + 5000).alias("Bonus_Salary"))
print(df_bonus)
```
