```python
# New dataset with employee details
import matplotlib.pyplot as plt
import polars as pl

df = pl.DataFrame({
    "Employee": ["Aman", "Rahul", "Priya", "Neha", "Rohit"],
    "Age": [22, 25, 20, 28, 24],
    "Department": ["IT", "HR", "IT", "Sales", "HR"],
    "Salary": [40000, 60000, 35000, 80000, 50000]
})

print(df)

it_dept = df.filter(pl.col("Department") == "IT")
print(it_dept)

high_salary = df.filter(pl.col("Salary") >= 50000)
print(high_salary)

df_bonus = df.with_columns((pl.col("Salary") + 5000).alias("Bonus_Salary"))
print(df_bonus)
```

```python
import polars as pl
import matplotlib.pyplot as plt

# 1. Data Create kiya
df = pl.DataFrame({
    "Employee": ["Aman", "Rahul", "Priya", "Neha", "Rohit"],
    "Age": [22, 25, 20, 28, 24],
    "Department": ["IT", "HR", "IT", "Sales", "HR"],
    "Salary": [40000, 60000, 35000, 80000, 50000]
})

# 2. Polars se data extract kiya (Graphing ke liye)
# .to_numpy() use karna sabse best hota hai AI Engineering mein
names = df["Employee"].to_numpy()
salaries = df["Salary"].to_numpy()

# 3. Graph Design
plt.figure(figsize=(10, 6))
plt.bar(names, salaries, color='purple', alpha=0.7)

# 4. Labels aur Title
plt.title("Salary Distribution by Employee")
plt.xlabel("Employee Name")
plt.ylabel("Salary")

# 5. Graph Display
plt.show()

# 6. Polars ka Extra Analysis: Average Salary by Department
dept_avg = df.group_by("Department").agg(pl.col("Salary").mean())
print("Average Salary by Department:")
print(dept_avg)
```
