```python
import numpy as np
import polars as pl

# 1. Python Function (Logic)
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"

# 2. NumPy (Vectorization)
arr = np.arange(10)
squared = arr ** 2
filtered = squared[squared > 25]

# 3. Polars (Basics)
df = pl.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
df = df.with_columns((pl.col("A") + pl.col("B")).alias("C"))

# Outputs
print("Divide result:", safe_divide(10, 2))
print("Filtered NumPy:", filtered)
print("Polars DF:\n", df)
```

```python
# student marks and analysis
import numpy as np 
import polars as pl

df = pl.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
data = {
    "Name": ["Lucky", "Aman", "Rahul", "Priya"],
    "Maths": [90, 30, 56, 95],
    "Science": [88, 65, 80, 45],
    "English": [23, 75, 78, 90]
}
df = pl.DataFrame(data)
print("Student Marks DataFrame:\n", df)
df = df.with_columns(
    (
        (pl.col("Maths") +
         pl.col("Science") +
         pl.col("English")) / 3
    ).alias("Average")
)
print("DataFrame with Average Marks:\n", df)
topper = df.sort("Average", descending=True)

print(topper)
```
