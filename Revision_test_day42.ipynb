```python
import numpy as np
import polars as pl

# 1. Python Function (Logic)
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"

# 2. NumPy (Vectorization)
arr = np.arange(10)
squared = arr ** 2
filtered = squared[squared > 25]

# 3. Polars (Basics)
df = pl.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
df = df.with_columns((pl.col("A") + pl.col("B")).alias("C"))

# Outputs
print("Divide result:", safe_divide(10, 2))
print("Filtered NumPy:", filtered)
print("Polars DF:\n", df)
```

```python
# student marks and analysis
import numpy as np 
import polars as pl

df = pl.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
data = {
    "Name": ["Lucky", "Aman", "Rahul", "Priya"],
    "Maths": [90, 30, 56, 95],
    "Science": [88, 65, 80, 45],
    "English": [23, 75, 78, 90]
}
df = pl.DataFrame(data)
print("Student Marks DataFrame:\n", df)
df = df.with_columns(
    (
        (pl.col("Maths") +
         pl.col("Science") +
         pl.col("English")) / 3
    ).alias("Average")
)
print("DataFrame with Average Marks:\n", df)
topper = df.sort("Average", descending=True)

print(topper)
```

```python
# e-commerce sales analytics(Numpy + polars)
import numpy as np 
import polars as pl

np.random.seed(42)

customers = [
    "Lucky", "Aman", "Rahul", "Priya", "Neha",
    "Karan", "Simran", "Arjun", "Riya", "Vikas"
]

categories = [
    "Electronics",
    "Fashion",
    "Beauty",
    "Home",
    "Sports"
]

products = [
    "Laptop", "Shoes", "Perfume", "Sofa", "Football",
    "Camera", "Jacket", "Lipstick", "Table", "Watch"
]

data = {
    "Order_ID": np.arange(1001, 1051),

    "Customer": np.random.choice(customers, 50),

    "Category": np.random.choice(categories, 50),

    "Product": np.random.choice(products, 50),

    "Price": np.random.randint(500, 80000, 50),

    "Quantity": np.random.randint(1, 5, 50),

    "Discount_Percent": np.random.randint(5, 30, 50),

    "Rating": np.round(
        np.random.uniform(3.5, 5.0, 50),
        1
    )
}

df = pl.DataFrame(data)

# Revenue
df = df.with_columns(
    (
        pl.col("Price") * pl.col("Quantity")
    ).alias("Revenue")
)

# Discount Amount
df = df.with_columns(
    (
        pl.col("Revenue") *
        pl.col("Discount_Percent") / 100
    ).alias("Discount_Amount")
)

# Final Revenue
df = df.with_columns(
    (
        pl.col("Revenue") -
        pl.col("Discount_Amount")
    ).alias("Final_Revenue")
)

# Profit
df = df.with_columns(
    (
        pl.col("Final_Revenue") * 0.25
    ).alias("Profit")
)

df = pl.DataFrame(data)
print("E-commerce Sales DataFrame:\n", df)
```

```python
df = df.with_columns(
    (pl.col("Price") * pl.col("Quantity")).alias("Total_Sales")
)
print("DataFrame with total sales:\n", df)
```
